## Analysing the Analysers
* The role of ‘data scientist’ is poorly defined, both by recruiters and those already acting in such a role. This can be attributed by the sudden emergence of the profession brought on by the proliferation of data available to organisations and the the realisation of technologies capable of analysing such large data sets. The lack of consensus surrounding what is expected of a data scientist results in undue waste of time and expense for both companies and candidates during the recruitment process. 

* Harris, Patrick & Vaisman set out to offer clarity on this issue. Using a five-page survey offered to data practitioners, the authors enquired about respondents skills, experiences, self-identification and web presence. From the questions asked, the authors engaged in cluster analysis to identify 4 sub-groups of data practioners, each correlated with a set of skills (of which had also undergone cluster analysis). 

* The 4 sub-groups as identified by Harris, Patrick & Vaisman were:
 * Data Business people - most focused on the organisation and how data yields profits. Many were entrepreneurs and had lead teams of employees in the past. 
 * Data Creatives - the ‘jack-of-all-trades’. These individuals had the broadest skill set as they had the most variable in the way they ranked their data skills. 
 * Data Developers - Focused on the technical problem of managing data and were the wrote the most code of the 4 ID groups. The majority of data developers had enginneriring or computer science backgrounds. 
 * Data Researcher - Came from an academic background with the majority having published in peer-reviewed journals. 

* The findings here offer partial support for the ’T-shaped data scientist’ (as defined by Mason & Wiggins, 2010). The idea is that the most successful data scientists are those with a breadth of skills available to offer analysis and insights from data as well as a deeper expertise in at least one aspect of data science. The strongest support for the ’T-shaped model’ can be seen for Data Businesspeople and, to a lesser extent, Data Researchers. 

* Overall, the findings from this report is a strong model for candidates to communicate their expertise and for organisations both when seeking to recruit to fulfil new positions and when training existing data professionals

## Teaching Data Science and Storytelling

Storytelling, according  Joe Blitzstein - Professor of Practice of Statistics at Harvard - is an essential skill for aspiring data scientists:

>  *I think there is a real art to getting interpretable results and then communicating those results, 
>  especially in the age of big data where you have thousands of variables.*

Blitzstein views effective story-telling skills as the bridge that links complex data analysis to meaningful results that mobilise change within organisations. He offers the example of data conferences where presenters rush through their slides, but fail to communicate the consequence of their findings. Indeed, in such a collaborative industry, data scientists are required to work in teams of people with diverse backgrounds. Those who succeed in this profession need the ability to put themselves in the shoes of their audience, to convey complex ideas and findings into terms that align with their skills and interests. 

The ability to communicate complex ideas simply benefits the individual data scientist as well:
> *When you can’t explain something clearly, it’s a sign that you haven’t thought it through fully 
> yourself*

A data scientist who can not convey information effectively may need to reassess their understanding of their own research. 

When asked to offer advice for those interested in studying data science, Blitzstein says that one should constantly think about the ‘why’ behind the research. Why is this data set interesting? Why is it relevant?

In summary, effective story-telling is an immuatable skill for all current and aspiring data scientists. When this ability is lacking, individuals may find that the impact of their projects are left unappreciated and innocuous.
